```python
pd.set_option('display.max_columns', 60)
complaints_df = pd.read_csv('../data/311-service-requests.csv', low_memory=False)
complaints_df
```

```python
complaint_counts = complaints_df['Complaint Type'].value_counts()
complaint_counts[:10]
```

```python
complaint_counts.plot(kind="bar")
plt.show()
```

```python
noise_complaints = complaints_df[complaints_df['Complaint Type'] == 'Noise - Street/Sidewalk']
noise_complaints.head()
```

```python
complaints = complaints_df
```

```python
is_noise = complaints['Complaint Type'] == "Noise - Street/Sidewalk"
in_brooklyn = complaints['Borough'] == "BROOKLYN"
complaints[is_noise & in_brooklyn][:5]
```

```python
is_noise = complaints['Complaint Type'] == "Noise - Street/Sidewalk"
noise_complaints = complaints[is_noise]
x = noise_complaints['Borough'].value_counts()
```

```python
x.plot(kind='bar')
plt.show()
```

```python
noise_complaint_counts = noise_complaints['Borough'].value_counts()
complaint_counts = complaints['Borough'].value_counts()
noise_complaint_counts / complaint_counts.astype(float)
(noise_complaint_counts / complaint_counts.astype(float)).plot(kind='bar')
plt.show()
```
